## Background

Safe Haven Security (https://www.mysafehaven.com), the company moving forward, is the largest ADT sub-dealer accounting for 100,000+ home security sales and installations.


One of the larger sales channels of the company receives thousands of leads per day from sites like Zillow, RedFin, Realtor, and MLS for newly sold homes--which are all treated as leads.
With this pool of new leads, the Leads team filters and handpicks leads to send to the boots on the ground department with offices and reps across 48 of the states.

> In reality, the department tries to touch a lead about 2 weeks after homeowner closed on the property.


> *Touching a lead is the act of making contact with the lead in some capacity. For this department, touching entails driving to and literally knocking on the lead's door.* 


This department drives to each lead/address they receive to try and sell security at the door to the new home owner, and though across the 48 states there are around 1,000 sales reps in this department, the amount of leads they can "touch" is finite. 


> The boots on the ground sales force "Touches" about 12,000 leads per day across the United States at about 12 leads per rep per day.

In 2021, the department touched over 3 million leads across the United States; however, as home sales in the United States in 2021 nearly hit 7 million, there were about 4 million leads ignored simply due to capacity. 


> The leads team will make determinations on whether or not the salability of a lead warrants a touch based on market/zipcode, reps available in the market, value of the home, and gut feelings. 

Safe Haven Security would like to improve the picking of leads to touch prioritized on the salability of the leadss.

## Project Description

The SSNH project aims to determine a "Salability" score of a lead based on an assortment of data points, and with this score, the company will be able to better prioritize touching leads by market by score. 

### Primary Use Case (PUC)

As the head of the Leads team, I want to take a csv of leads and pump them through the SSNH, providing the SSNH score to each lead. I will then group by market and prioritize leads within the markect by the SSNH score. 

### Raw Data Sources

The company recently make large breakthroughs getting their data from their CRMs, HR System, Vendor EDI Sources, and Affiliate partners into a enterprise data core made up by a data lake and a data warehouse.

The data to be used for linnear regression will mainly come from the company's core CRM that contains the contractual data (i.e. whether or not the lead made it to install) and an auxiliary CRM that houses additional data like the value of the property. 

Then of course the data to feed the ssnh will come from feature engineering the daily leads intake from Zillow, RedFin, Realtor, and the MLS. 

## Performance Metric
As the head of the Leads team, I want to be able to send a CSV file of new leads through the SSNH to provide a salability score to all of the leads. With this score, I will group by market and prioritize by the ssnh score.

To determine the performance and thus value of the machine learning project to score salabity for the automation of selecting leads to pursue, the company can track monthly or even weekly **closing rate** such that **closing rate** is *the amount of installs divided by the amount of leads touched in the time period after implementing the scoring*.

### A Little Goes A Long Way

In 2021, the company had a **closing rate** of about 2% on 3 million leads or 60 thousands installations, accounting for about $120 million dollars of gross revenue.

> If SSNH scoring leads to a 5% increase to the **closing rate** of the department, then it would directly create an additional $6 million dollars of gross revenue. 

> If SSNH scoring leads to a 25% increase to the **closing rate** of the department, then it would directly create an additional $30 million dollars of gross revenue. 

> If SSNH scoring leads to a 50% increase to the **closing rate** of the department, then it would directly create an additional $60 million dollars of gross revenue.
